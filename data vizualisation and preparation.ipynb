```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
df02 = pd.read_csv (r'C:\Users\PROXY INFO\Downloads\titanic-passengers (1).csv', sep = ';')
encaa=LabelEncoder()
df02['Survived']=encaa.fit_transform(df02['Survived'])
df02['Sex']=encaa.fit_transform(df02['Sex'])
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( df02 ) 
```

```python
df03 = df02.dropna(axis=0, subset=['Age'])
def f(age):
    
        if age['Age']<10:
            val=0
        elif 10<=age['Age']<20:
            val=10
        elif 20<=age['Age']<30:
            val=20
        elif 30<=age['Age']<40:
            val=30
        elif 40<=age['Age']<50:
            val=40
        elif 50<=age['Age']<60:
            val=50
        elif 60<=age['Age']<70:
            val=60
        else:
            val=70   
        return val
df03['interval age'] = df03.apply(f, axis=1)
df03
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
df01 = pd.read_csv (r'C:\Users\Pentest\Downloads\titanic-passengers.csv', sep = ';')
df01
```

```python
st = df01['Name'].str.split(',', n = 1, expand = True)
st
print(type(st))
```

```python
print(st[0].value_counts())
```

```python
st = st[1].str.split('.', n = 1, expand = True)
st
```

```python
print(st[0].value_counts())
```

```python
df02 = pd.concat([df01, st], axis=1)
df02
```

```python
df02.rename(columns = {0:'title'}, inplace = True) 
df02
```

```python
df03=df02.drop([ "Ticket","Cabin","Embarked","Name",1],axis =1)
df03
```

```python
grid= sns.FacetGrid(df03, row='Survived', col='Sex', height=2.2, aspect=1.6)
grid.map(sns.barplot, 'Age', 'title', alpha=0.5,ci=None)
grid.add_legend()
```

```python
Title_Dictionary = {

                    "Capt":"Officer",

                    "Col":"Officer",

                    "Major":"Officer",

                      "Dr":"Officer",

                    "Rev":"Officer",

                    "Jonkheer":"Royalty",

                    "Don":"Royalty",

                    "Sir" :"Royalty",

                   "Lady" :"Royalty",

                  "the Countess":"Royalty",

                    "Dona":"Royalty",

                    "Mme":"Miss",

                    "Mlle":"Miss",

                    "Miss" :"Miss",

                    "Ms":"Mrs",

                    "Mr" :"Mrs",

                    "Mrs" :"Mrs",

                    "Master" :"Master"

                    }
df03['title']=df03['title'].str.strip()
df03["title"].replace(Title_Dictionary, inplace=True)
df03
```

```python
print(df03['title'].value_counts())
```

```python
grid= sns.FacetGrid(df03, row='Survived', col='Sex', height=2.2, aspect=1.6)
grid.map(sns.barplot, 'title', 'Age', alpha=0.5,ci=None)
grid.add_legend()
```

```python
df03['Familiysize'] = df03['SibSp'] + df03['Parch']
df03
```

```python
grid= sns.FacetGrid(df03, row='Survived', col='Sex', height=2.6, aspect=1.6)
grid.map(sns.barplot, 'Familiysize', 'title', alpha=0.5,ci=None)
grid.add_legend()
```
